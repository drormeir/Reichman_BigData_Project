```python
import numpy as np
import threading
from MockObjectStorage import MockObjectStorage
```

```python
objectStorage = MockObjectStorage() # can also set param: MaxCachedFiles=0


print("Creating data...")

files = []
for i in range(10):
    filename = "File" + str(i) + ".csv"
    objectStorage.createObject(filename,"Data of " + filename)
    files.append(filename)

objectStorage.flush()
print("Data is created.")
```

```python
datas = objectStorage.readObject(files,[str])
print(datas)
```

```python
def thread_func(files, threadID, times):
    print("Starting thread",threadID)
    for t in range(times):
        filename = np.random.choice(files)
        print("Thread {} execution number: {} will read the file: {}".format(threadID,t+1,filename))
        data = objectStorage.readObject(filename,str)
        print("Thread {} reads the result: {}".format(threadID,data))
    print("Ending thread",threadID)

        
times = 100
threads = 20
threads_vec = [threading.Thread(target=thread_func, args=(files,i,times)) for i in range(threads)]
```

```python
print("Starting",threads,"threads to run for",times,"times each")
for t in threads_vec:
    t.start()
print("Wait for threads to finish")
for t in threads_vec:
    t.join()
print("All threads completed") 
```

```python
objectStorage.flush()

objectStorage.deleteObject(files)
```
