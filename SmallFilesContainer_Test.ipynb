```python
from SmallFilesContainer import SmallFilesContainer
import numpy as np
from tqdm import tqdm
```

```python
smallFilesContainer = SmallFilesContainer("SmallFilesTest.csv")
```

```python
print("Creating small files data base")

firstname  = ['John', 'Dana', 'Scott', 'Marc', 'Steven', 'Michael', 'Albert', 'Johanna']
secondname = ['Smith', 'Brown', 'Miller', 'Watson', 'Bain']
city       = ['NewYork', 'Haifa', 'Munchen', 'London', 'PaloAlto',  'TelAviv', 'Kiel', 'Hamburg']

for ind_file in tqdm(range(10)):
    filename = "File_{:05d}.csv".format(ind_file)
    pop = np.random.randint(low=5,high=101)
    first = np.random.choice(a=firstname, size=pop)
    second = np.random.choice(a=secondname, size=pop)
    cit = np.random.choice(a=city, size=pop)
    age = np.random.choice(a=50,size=pop) + 18
    child = np.random.choice(a=5,size=pop)
    rows = [first[p] + "," + second[p] + "," +  cit[p] + "," +str(age[p]) + "," + str(child[p]) for p in range(pop)]
    smallFilesContainer.createNewFile(filename,rows,deleteExist=True)

smallFilesContainer.flush(True)
```
